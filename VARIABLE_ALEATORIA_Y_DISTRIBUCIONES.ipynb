#**PARTE A: Distribución Marginal y Condicional**

Considere el siguiente problema:

En una empresa de café llamado San Alberto, famoso por su proceso de quinta selección de grano de café, tiene un lote de café para ser vendido en los próximos tres meses. Hay 2 tipos de café que producen en la finca: mediano y oscuro, para ser vendidos a Cartagena, Bogotá y Medellín. En dicho lote, se utiliza 30 granos de café para hacer selección de los mejores granos, de los cuales hay 6 granos de quinta selección, 16 granos de cuarta selección, y 8 granos de tercera selección. Si se selecciona al azar 2 granos de café, cualés son las probabilidades que salga granos de tercera, cuarta y quinta selección? ¿Cuáles son las probabilidades que de un grano de tercera selección salga un grano de quinta selección y de un grano de cuarta selección salga un grano de quinta selección?

Para este problema, considere:

$X$ = Número de libras de café de cuarta selección

$Y$ = Número de libras de café de quinta selección

$$X + Y \leq 30$$

```
from math import factorial

def bin_comb(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))
```

```
import numpy as np
import pandas as pd

Total_granos_cafe = 30
granos_tercera_seleccion = 8
granos_cuarta_seleccion = 16
granos_quinta_seleccion = 6
granos_cafe_elegidos = 2
```

```
#caso 1 x=0 y=0 


Numero_total_selecciones_cafe = bin_comb(Total_granos_cafe, granos_cafe_elegidos)
Numero_eventos_cafe_tercera = bin_comb(granos_tercera_seleccion, granos_cafe_elegidos)
print("Número total de selecciones posibles de cafe = ", Numero_total_selecciones_cafe)
print("Número de eventos que salga cafe de tercera generación = ", Numero_eventos_cafe_tercera)
Prob_X0_Y0 = Numero_eventos_cafe_tercera / Numero_total_selecciones_cafe
print('Probabilidad de que los dos granos de cafe sean de tercera generación = {:.5f}'. format(Prob_X0_Y0))
```

```
#caso 2 x=0 y= 1 

Total_granos_cafe = 30
granos_tercera_seleccion = 8
granos_quinta_seleccion = 6
granos_cafe_elegidos = 1

Numero_eventos_con_quinta = bin_comb(granos_quinta_seleccion, granos_cafe_elegidos)
Numero_eventos_cafe_tercera_con_cafe_quinta = bin_comb(granos_tercera_seleccion, granos_cafe_elegidos)
Prob_X0_Y1 = (Numero_eventos_con_quinta * Numero_eventos_cafe_tercera_con_cafe_quinta) / Numero_total_selecciones_cafe
print('Probabilidad de elegir un cafe de tercera generación y un cafe de quinta generación = {:.5f}'. format(Prob_X0_Y1))
```

```
#caso 3 x=0 y=2 

Total_granos_cafe = 30
granos_quinta_seleccion = 6
granos_cafe_elegidos = 2

Numero_eventos_cafe_dos_quinta = bin_comb(granos_quinta_seleccion, granos_cafe_elegidos)
Prob_X0_Y2 = Numero_eventos_cafe_dos_quinta / Numero_total_selecciones_cafe
print('Probabilidad que los dos cafes sean de quinta generación = {:.5f}'. format(Prob_X0_Y2))
```

```
#caso 4 x=1  y=0 

Total_granos_cafe = 30
granos_tercera_seleccion = 8
granos_cuarta_seleccion = 16
granos_cafe_elegidos = 1

Numero_eventos_un_cafe_cuarta = bin_comb(granos_cuarta_seleccion, granos_cafe_elegidos)
Numero_eventos_cafe_tercera_con_cuarta = bin_comb(granos_tercera_seleccion, granos_cafe_elegidos)
Prob_X1_Y0 = (Numero_eventos_un_cafe_cuarta * Numero_eventos_cafe_tercera_con_cuarta ) / Numero_total_selecciones_cafe 
print('Probabilidad de selccionar un cafe tercera generación y uno de cuarta generación = {:.5f}'. format(Prob_X1_Y0))
```

```
#caso 5 x=2 y=0 

Total_granos_cafe = 30
granos_tercera_seleccion = 8
granos_cuarta_seleccion = 16
granos_cafe_elegidos = 2

Numero_eventos_cafe_con_dos_cuarta = bin_comb(granos_cuarta_seleccion, granos_cafe_elegidos)
Prob_X2_Y0 = Numero_eventos_cafe_con_dos_cuarta / Numero_total_selecciones_cafe
print('Probabilidad de seleccionar con dos cafe de cuarta generación = {:.5f}'. format(Prob_X2_Y0))
```

```
# caso 6 x=1 y=1 

Total_granos_cafe = 30
granos_tercera_seleccion = 8
granos_cuarta_seleccion = 16
granos_quinta_seleccion = 6
granos_cafe_elegidos = 1

Numero_eventos_cafe_con_cuarta = bin_comb(granos_cuarta_seleccion, granos_cafe_elegidos)
Numero_eventos_cafe_con_quinta = bin_comb(granos_quinta_seleccion, granos_cafe_elegidos)
Prob_X1_Y1 = (Numero_eventos_cafe_con_cuarta * Numero_eventos_cafe_con_quinta) / Numero_total_selecciones_cafe
print('Probabilidad de seleccionar un cafe de cuarta y uno de quinta generacion = {:.5f}'. format(Prob_X1_Y1))
```

```
#Orgnizar en tabla

tabla_probabilidad= np.array([[Prob_X0_Y0, Prob_X1_Y0, Prob_X2_Y0],[Prob_X0_Y1, Prob_X1_Y1, 0],[Prob_X0_Y2,0,0]])
casos= ["0", "1", "2"]
datos= pd.DataFrame(data = tabla_probabilidad, columns = casos, index = casos)
display(datos)

##Conclusion> Hay mayor probabilidad que al elegir dos granos de cafe al azar, uno sea de cuarta generacion y otro de tercera generacion, 
#que en este caso seria el caso 4 correspondiente a x=1  y=0, con una probabilidad marginal de 0.294353
#A continuacion veremos las probabilidades de elegir un grano de cafe de tercera, cuarta y quinta generacion
print(f"la probabilidad de que un grano sea de tercer generacion es: {Prob_X1_Y0 + Prob_X2_Y0 }")
print(f"la probabilidad de que un grano sea de cuarta generacion es: {Prob_X1_Y0 + Prob_X0_Y0}")
print(f"la probabilidad de que un grano sea de quinta generacion es:{Prob_X2_Y0+ Prob_X1_Y0 + Prob_X0_Y0}" )
```

```
Suma_Prob_Filas = np.sum(tabla_probabilidad, axis=1)
Suma_Prob_Columnas = np.hstack((np.sum(tabla_probabilidad, axis=0),1))
print("Suma de probabilidades filas: \n",Suma_Prob_Filas)
print("Suma de probabilidades columnas: \n",Suma_Prob_Columnas)
```

```
#organizo en tabla 
Arreglo_Fila = np.array(Suma_Prob_Filas, copy=False, subok=True, ndmin=2).T
MAT = np.column_stack([tabla_probabilidad,Arreglo_Fila])
MATRIZ = np.vstack((MAT,Suma_Prob_Columnas))
Columnas = ['0', '1', '2', 'Suma Probabilidad']
Filas = ['0', '1', '2', 'Suma Probabilidad']
Matriz_Probabilidades = pd.DataFrame(data = MATRIZ, columns = Columnas, index = Filas)
display(Matriz_Probabilidades)
```

```
#¿Cuáles son las probabilidades que de un grano de tercera selección salga un grano de quinta selección?

#probabilidad de que un grano salga de 4 generacion, dado que el primero es de tercera generacion 
Prob_X1_cond_Y0 = Matriz_Probabilidades.iloc[0,1]/Matriz_Probabilidades.iloc[0,3]
print("Probabilidad de seleccionar un grano de cafe de cuarta generacion DADO que la otra es de tercera = {:.3f}".format(Prob_X1_cond_Y0))
```

#**PARTE H: Media vs Media Geométrica**

La media geométrica es una medida que que calcula la media de un conjunto de datos por medio del logaritmo natural de los valores:

$$\mu_{geometrica} = \left ( \prod_{n=1}^N x_n \right)^{\frac{1}{N}} = \exp \left( \frac{\sum_{n=1}^N ln(x_n)}{N}\right )$$

Según los estadísticos, la media geométrica mejora la localización de una distribución, es decir, que mejora el calculo de la media de la distribución. Sin embargo, esta medida es valida solo para valores de $x$ positivos y $x > 1$, para evitar valores indeterminados por el cálculo del logaritmo natural.
